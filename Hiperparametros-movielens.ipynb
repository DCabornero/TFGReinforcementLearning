# An√°lisis de epsilon-greedy

```python
from Analysis import Analysis
from EpsilonGreedy import EpsilonGreedy

an = Analysis(path='results',name='epsilon_ext')
lst = []
epsilon = [0.1,0.2,0.5]
alpha = [0.01,0.1,0.3]

for ep in epsilon:
    for al in alpha:
        bandit = EpsilonGreedy(epsilon=ep,alpha=al)
        bandit.name += ',epsilon:'+str(ep)+',alpha:'+str(al)
        bandit.read_csv('data/ratings.csv')
        lst.append(bandit)


an.execute(lst,1e6)
an.recall()
```

# An√°lisis general: no contextuales, 100K

```python
from Analysis import Analysis
from RandomBandit import RandomBandit
from EpsilonGreedy import EpsilonGreedy
from UCB import UCB
from CNAME import CNAME
from Exp3 import Exp3
from ThompsonSampling import ThompsonSampling
from GradientBandit import GradientBandit

bandits = []
bandits.append(RandomBandit())
bandits.append(EpsilonGreedy(epsilon=0.2,initial=0.3))
bandits.append(UCB(c=0.1))
bandits.append(CNAME(w=5))
bandits.append(GradientBandit(alpha=5))
bandits.append(Exp3(k=50))
bandits.append(ThompsonSampling(alpha=1,beta=1000))

an = Analysis(path='results',name='RecallNoContext100k_')
for bandit in bandits:
    bandit.read_csv('data/ratings.csv')

an.execute(bandits,1e5)
an.recall()
an.time()
an.gini()

print('Recall acumulado:')
for bandit in bandits:
    print(str(bandit)+': '+str(np.round(bandit.recall[-1],4)))

print('Tiempos:')
for bandit in bandits:
    print(str(bandit)+': '+str(datetime.timedelta(seconds=bandit.times)))

print('Gini:')
for bandit in bandits:
    print(str(bandit)+': '+str(np.round(bandit.gini(),3)))
```
