# Análisis general: no contextuales 100K

```python
from Analysis import Analysis
from RandomBandit import RandomBandit
from EpsilonGreedy import EpsilonGreedy
from UCB import UCB
from CNAME import CNAME
from Exp3 import Exp3
from ThompsonSampling import ThompsonSampling
from GradientBandit import GradientBandit

bandits = []
bandits.append(RandomBandit())
bandits.append(EpsilonGreedy(epsilon=0.2,initial=0))
bandits.append(UCB(c=0.1))
bandits.append(CNAME(w=2))
bandits.append(GradientBandit(alpha=2))
bandits.append(Exp3(k=50))
bandits.append(ThompsonSampling(alpha=1,beta=100))

an = Analysis(path='results',name='NoContext100K_')
for bandit in bandits:
    bandit.read_csv('data/cm100k/ratings.csv',**dict_names)

an.execute(bandits,1e5,gini_history=True)
an.recall()
an.gini()

print('Recall acumulado:')
for bandit in bandits:
    print(str(bandit)+': '+str(np.round(bandit.recall[-1],4)))

# print('Tiempos:')
# for bandit in bandits:
#     print(str(bandit)+': '+str(datetime.timedelta(seconds=bandit.times)))

print('Gini:')
for bandit in bandits:
    print(str(bandit)+': '+str(np.round(bandit.gini(),3)))
```
