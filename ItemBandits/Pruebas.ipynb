```python
from RandomBandit import RandomBandit

bandit = RandomBandit('ratings.csv')
rnd = bandit.run_epoch(epochs=100000)
```

```python
from EpsilonGreedy import EpsilonGreedy

bandit = EpsilonGreedy('ratings.csv',epsilon=0.1,alpha=0.1)
eps = bandit.run_epoch(epochs=100000)
```

```python
from ThompsonSampling import ThompsonSampling

bandit = ThompsonSampling('ratings.csv')
thomps = bandit.run_epoch(epochs=100000)
```

```python
from ThompsonSampling import ThompsonSampling
import numpy as np


bandit = ThompsonSampling('ratings.csv',func='Bernouilli')
ber = bandit.run_epoch(epochs=100000)
```

```python
from UCB import UCB

bandit = UCB('ratings.csv',c=0.1)
ucb = bandit.run_epoch(epochs=100000)
```

```python
import matplotlib.pyplot as plt

graficas = [rnd,eps,thomps,ber,ucb]
labels = ['Random','EpsilonGreedy','Th.(Beta)','Th.(Bernouilli)','UCB']

for i,graf in enumerate(graficas):
    plt.plot(graf[0],np.array(graf[1])*100,label=labels[i])
    
plt.xlabel('Épocas')
plt.ylabel('Recall acumulado (%)')

plt.legend()
plt.plot()
```
