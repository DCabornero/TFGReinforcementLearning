# Análisis de epsilon-greedy

```python
from Analysis import Analysis
from EpsilonGreedy import EpsilonGreedy

an = Analysis(path='results',name='epsilon')
lst = []
epsilon = [0.01,0.1,0.2,0.5]
for ep in epsilon:
    bandit = EpsilonGreedy(epsilon=ep)
    bandit.name += ',epsilon:'+str(ep)
    bandit.read_csv('data/ratings.csv')
    lst.append(bandit)


an.execute(lst,1e6)
an.recall()
```

```python
from Analysis import Analysis
from EpsilonGreedy import EpsilonGreedy

an = Analysis(path='results',name='epsilon_ext')
lst = []
epsilon = [0.1,0.2,0.5]
alpha = [0.01,0.1,0.3]

for ep in epsilon:
    for al in alpha:
        bandit = EpsilonGreedy(epsilon=ep,alpha=al)
        bandit.name += ',epsilon:'+str(ep)+',alpha:'+str(al)
        bandit.read_csv('data/ratings.csv')
        lst.append(bandit)


an.execute(lst,1e6)
an.recall()
```

```python
# EPSILON A 0.2
from Analysis import Analysis
from EpsilonGreedy import EpsilonGreedy

an = Analysis(path='results',name='epsilon_cold')
lst = []
inits = [0,0.1,0.2,0.3,0.5,1]
for initial in inits:
    bandit = EpsilonGreedy(epsilon=0.2,initial=initial)
    bandit.name += ',initial:'+str(initial)
    bandit.read_csv('data/ratings.csv')
    lst.append(bandit)


an.execute(lst,1e6)
an.recall()
```
